```python
import pandas as pd
import openpyxl as op
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("dados_matriculas.xlsx" )
print(df)
```

```python
# verificar o numero de matriculas mensais
plt.figure(figsize=(10,5)) # Define o tamanho do grafico a ser exibido
df["Matrículas"].plot(marker='s', linestyle='--', color='red') # Define o tipo de marcador, estilo de linha e cor
plt.title("Matrículas Mensais") # Define o título do gráfico
plt.xlabel("Mês")  # Define o rótulo do eixo x   
plt.ylabel("Número de Matrículas")  # Define o rótulo do eixo y
plt.grid(True)  # Adiciona uma grade ao gráfico
#plt.show()  # Exibe o gráfico
```

```python
# identificar os meses com menor e maior numero de matriculas 
max_matriculas = df["Matrículas"].max() # Encontra o valor máximo de matrículas 
#mes_max_matriculas = df[df["Matrículas"] == max_matriculas]["Mês"].values[0] # Encontra o mês correspondente ao valor máximo de matrículas  
mes_max_matriculas = df["Matrículas"].idxmax() # Encontra o índice do mês com o valor máximo de matrículas
print(f"Mes com maior numero de matriculas: {mes_max_matriculas}: {max_matriculas}") # Imprime o mês com mais matrículas  
#-------------------------------------------------------------------------------------------------------------------

min_matriculas = df["Matrículas"].min() # Encontra o valor mínimo de matrículas
mes_min_matriculas = df["Matrículas"].idxmin() # Encontra o índice do mês com o valor mínimo de matrículas
print(f"Mes com menor numero de matriculas: {mes_min_matriculas}: {min_matriculas}") # Imprime o mês com menos matrículas   

#media de matriculas
media_matriculas = df["Matrículas"].mean() # Calcula a média de matrículas  



#min_matriculas = df["Matrículas"].min() # Encontra o valor mínimo de matrículas 
#mes_min_matriculas = df["Matrículas"].idxmin() # Encontra o índice do mês com o valor mínimo de matrículas
#print(f"Máximo de matrículas: {max_matriculas}") # Imprime o valor máximo de matrículas
#print(f"Mínimo de matrículas: {min_matriculas}") # Imprime o valor mínimo de matrículas

#print(f"Mês com mais matrículas: {mes_max_matriculas}") # Imprime o mês com mais matrículas
#print(f"Mês com menos matrículas: {mes_min_matriculas}") # Imprime o mês com menos matrículas
```

```python
# pltar grafico geral
plt.figure(figsize=(10,5)) # Define o tamanho do grafico a ser exibido
df["Matrículas"].plot(marker='o', linestyle='-', color='red', label='Matrículas', mfc='green', markersize="10") # Define o tipo de marcador, estilo de linha e cor
plt.title("Matrículas Mensais") # Define o título do gráfico

#adicionar linha de info para max e min de matriculas
plt.axhline(y=max_matriculas, color='green', linestyle='--', linewidth=2,label=f'Máximo: {max_matriculas}') # Adiciona uma linha horizontal para o valor máximo de matrículas
plt.axhline(y=min_matriculas, color='blue', linestyle='--', linewidth=2,label=f'Mínimo: {min_matriculas}') # Adiciona uma linha horizontal para o valor mínimo de matrículas
plt.axhline(y=media_matriculas, color='orange', linestyle='--', linewidth=2,label=f'Média: {media_matriculas:.2f}') # Adiciona uma linha horizontal para a média de matrículas
plt.legend() # Exibe a legenda do gráfico
plt.title("Matrículas Mensais") # Define o título do gráfico    
plt.xlabel("Mês")  # Define o rótulo do eixo x 
plt.ylabel("Número de Matrículas")  # Define o rótulo do eixo y 
plt.grid(True)  # Adiciona uma grade ao gráfico 
```

```python
#tendencia de matriculas 
#calcular a media de matriculas por trimestre
#QE trimestre
#QS semestre
#A anual
#D diario
#W semanal
#M mensal
#H horario
```

```python
# passo 1 - converter para datetime
df["Mês"] = pd.to_datetime(df["Mês"]) # Converte a coluna "Mês" para o formato datetime, especificando o formato de entrada

# passo 2 - definir a coluna "Mês" como índice do DataFrame 
df.set_index("Mês", inplace=True) # Define a coluna "Mês" como índice do DataFrame, permitindo operações de série temporal

# passo 3 - calcular a média de matrículas por trimestre   
media_trimestral = df["Matrículas"].resample("QE").mean() # Calcula a média de matrículas por trimestre usando o método resample com frequência "Q" (trimestral) e a função mean()
print(media_trimestral) # Imprime a média de matrículas por trimestre
```

```python
#plotar grafico de barras - media trimestral
plt.figure(figsize=(10,5)) # Define o tamanho do grafico a ser exibido
media_trimestral.plot(kind='bar', color='skyblue') # Plota um gráfico de barras para a média trimestral de matrículas
plt.title("Média Trimestral de Matrículas") # Define o título do gráfico    
plt.xlabel("Trimestre") # Define o rótulo do eixo x
plt.ylabel("Média de Matrículas") # Define o rótulo do eixo y
plt.xticks(rotation=0) # Rotaciona os rótulos do eixo x em 45 graus
plt.tight_layout() # Ajusta o layout do gráfico para evitar sobreposição
plt.grid(axis='y') # Adiciona uma grade apenas no eixo y
plt.show() # Exibe o gráfico    
```
