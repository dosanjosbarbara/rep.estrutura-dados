```python
# passo 1 - converter para datetime
df["Mês"] = pd.to_datetime(df["Mês"]) # Converte a coluna "Mês" para o formato datetime, especificando o formato de entrada

# passo 2 - definir a coluna "Mês" como índice do DataFrame 
df.set_index("Mês", inplace=True) # Define a coluna "Mês" como índice do DataFrame, permitindo operações de série temporal

# passo 3 - calcular a média de matrículas por trimestre   
media_trimestral = df["Matrículas"].resample("QE").mean() # Calcula a média de matrículas por trimestre usando o método resample com frequência "Q" (trimestral) e a função mean()
print(media_trimestral) # Imprime a média de matrículas por trimestre
```

```python
#plotar grafico de barras - media trimestral
plt.figure(figsize=(10,5)) # Define o tamanho do grafico a ser exibido
media_trimestral.plot(kind='bar', color='skyblue') # Plota um gráfico de barras para a média trimestral de matrículas
plt.title("Média Trimestral de Matrículas") # Define o título do gráfico    
plt.xlabel("Trimestre") # Define o rótulo do eixo x
plt.ylabel("Média de Matrículas") # Define o rótulo do eixo y
plt.xticks(rotation=0) # Rotaciona os rótulos do eixo x em 45 graus
plt.tight_layout() # Ajusta o layout do gráfico para evitar sobreposição
plt.grid(axis='y') # Adiciona uma grade apenas no eixo y
plt.show() # Exibe o gráfico    
```

```python
#seaborn para plootar grafico de linha com variacao percentual mensal
import pandas as pd
import seaborn as sns
import jinja2
#criar tabela
 
variacao_percentual = pd.DataFrame({
    "Matrículas":df["Matrículas"],
    "Variação Percentual (%)": variacao_percentual
}).round(2)
 
# Exibir tabela formatada
 
def colorir_variacao(val):
  if val > 0:
    color = 'green' #verde para crescimento
  elif val < 0:
    color = 'red' #vermelho para queda
  else:
    color = 'purple' #roxo para valores sem variacao
  return f'color: {color}'
 
variacao_percentual.style.map(colorir_variacao, subset = ["Variação Percentual (%)"])
```

```python
#3 identificar os meses com maior crescimento e maior queda
maior_crescimento = variacao_percentual["Variação Percentual (%)"].idxmax() # Encontra o valor máximo de variação percentual
maior_queda = variacao_percentual["Variação Percentual (%)"].idxmin() # Encontra o valor mínimo de variação percentual
mes_maior_crescimento = variacao_percentual["Variação Percentual (%)"].loc[maior_crescimento] # Obtém o valor da variação percentual para o mês com maior crescimento
mes_maior_queda = variacao_percentual["Variação Percentual (%)"].loc[maior_queda]
print(f"Mes com maior crescimento: {maior_crescimento}: {mes_maior_crescimento:.2f}%") # Imprime o mês com maior crescimento e o valor da variação percentual
print(f"Mes com maior queda: {maior_queda}: {mes_maior_queda:.2f}%") # Imprime o mês com maior queda e o valor da variação percentual
```

```python
semestral = df["Matrículas"].resample("6ME").sum() # Calcula a soma de matrículas por semestre usando o método resample com frequência "6ME" (semestre) e a função sum()
semestral.indez = semestral.index.strftime('%Y-%m') # Formata o índice do DataFrame semestral para exibir apenas o ano e o mês no formato "YYYY-MM"
semestral.plot(kind="bar", color='skyblue') # Plota um gráfico de barras para a soma de matrículas por semestre
plt.title("Soma de Matrículas por Semestre") # Define o título do gráfico   
plt.xlabel("Semestre") # Define o rótulo do eixo x
plt.ylabel("Soma de Matrículas") # Define o rótulo do eixo y
plt.xticks(rotation=45) # Rotaciona os rótulos do eixo x em
plt.tight_layout() # Ajusta o layout do gráfico para evitar sobreposição
plt.show() # Exibe o gráfico
```

```python
media = df["Matrículas"].mean() # Calcula a média de matrículas 
mediana = df["Matrículas"].median() # Calcula a mediana de matrículas   
moda = df["Matrículas"].mode() # Calcula a moda de matrículas
desvio = df["Matrículas"].std() # Calcula o desvio padrão de matrículas
coef_var = (desvio / media) * 100 # Calcula o coeficiente de variação de matrículas

print(f"Media: {media:.2f}") # Imprime a média de matrículas formatada com 2 casas decimais
print(f"Mediana: {mediana:.2f}") # Imprime a mediana de matrículas formatada com 2 casas decimais
print(f"Moda: {moda.values}") # Imprime a moda de matrículas
print(f"Desvio Padrão: {desvio:.2f}") # Imprime o desvio padrão de matrículas formatado com 2 casas decimais
print(f"Coeficiente de Variação: {coef_var:.2f}%") # Imprime o coeficiente de variação de matrículas formatado com 2 casas decimais e símbolo de porcentagem

if media > mediana:
    print("A distribuição é assimétrica a direita (positiva).") # Imprime que a distribuição é assimétrica positiva se a média for maior que a mediana
elif media < mediana:
    print("A distribuição é assimétrica a esquerda (negativa).") # Imprime que a distribuição é assimétrica negativa se a média for menor que a mediana
else:
    print("A distribuição é simétrica.") # Imprime que a distribuição é simétrica se a média for igual à mediana    
   
```
