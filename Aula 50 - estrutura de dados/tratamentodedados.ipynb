```python
#carregando o arquivo
import pandas as pd

df = pd.read_excel("dados_ausentes.xlsx")
print(df)
```

```python
#verificar os dados ausentes
print(df.isnull())
```

```python
#contar o numero de dados ausentes
print(df.isnull().sum())    
```

```python
#tratando os dados ausentes, excluindo as linhas com dados ausentes
df_sem_vazio = df.dropna() #necessario criar uma nova variavel para armazenar o resultado - dropna() nao altera o dataframe original e exclui os dados nan
print(df_sem_vazio)
```

```python
#tratar os dados nan - substituindo os dados ausentes por um valor especifico
#cidade vazio = desconhecido
#df_cidade_desconecido = df.fillna({"cidade": "desconecido"}) // # necessario criar uma nova variavel para armazenar o resultado - fillna() nao altera o dataframe original e substitui os dados nan pelo valor especificado ({"cidade": "desconecido"})

df["Cidade"] = df["Cidade"].fillna("Desconecido", inplace=False) #substitui os dados nan da coluna cidade por "desconecido" e altera o dataframe original (inplace=True)
print(df) # retorna o dataframe com os dados nan da coluna cidade substituidos por "desconecido" e os outros dados nan permanecem inalterados




```

```python
#tratando os dados ausentes, dados salario
df["Salário"] = df["Salário"].fillna(0, inplace=False) #substitui os dados nan da coluna salario por 0 e altera o dataframe original (inplace=True) 
print(df) # retorna o dataframe com os dados nan da coluna salario substituidos por 0 e os outros dados nan permanecem inalterados
```

```python
#df["Idade"] = df["Idade"].fillna(0, inplace=False)
#print(df)

#alterando a coluna de dado idade para a media da coluna idade
df["Idade"] = df["Idade"].fillna(df["Idade"].mean(), inplace=False) #calcula a media da coluna idade   
#df["Idade"] = df["Idade"].fillna(media_idade, inplace=False) #substitui os dados nan da coluna idade pela media da coluna idade e altera o dataframe original (inplace=True)
print(df) # retorna o dataframe com os dados nan da coluna idade substituidos pela media da 
```

```python
#corrigir erros de formatacao (normailizacao de dados) - padronizar os dados    
df["Cidade"] = df["Cidade"].str.lower() #padroniza os dados da coluna cidade para minusculo
```

```python
#corrigir erros de formatacao (normailizacao de dados) - padronizar os dados
df["Nome"] = df["Nome"].str.capitalize() #normaliza os dados da coluna nome para a primeira letra maiuscula e as outras minusculas
df["Cidade"] = df["Cidade"].str.title() #normaliza os dados da coluna cidade para a primeira letra de cada palavra maiuscula e as outras minusculas
print(df) # retorna o dataframe com os dados da coluna nome padronizados    
```

```python
#padronizacao de nomemclatura 
# replace = substitui valores especificos por outros valores especificos
df["Cidade"] = df["Cidade"].str.upper().replace({"sao palo": "SP",
                                                 "Sao Paulo": "SP",
                                                 "são paulo": "SP",
                                                 "SÃO PAULO ": "SP",
                                                 "RIO DE JANEIRO": "RJ",
                                                 "rio de janeiro": "RJ"}) #substitui os dados da coluna cidade "São Paulo" por "SP"
print(df) # retorna o dataframe com os dados da coluna cidade padronizados para "SP" e "RJ"
```

```python
#identificar tipos de dados 
#tipo string

coluna_string = df.select_dtypes(include=["string"]).columns.to_list() # seleciona as colunas do dataframe que possuem o tipo de dado string e retorna uma lista com os nomes dessas colunas
print(coluna_string) # retorna a lista com os nomes das colunas do dataframe que possuem o tipo de dado string
```

```python
#trocar tipo e dado de float para int
df["Idade"] = df["Idade"].astype("int64") # converte a coluna idade de float para int   
print(df)
```

```python
#pegar o valor de cima e substituir o valor de baixo
#df["Salário"] = df["Salário"].fillna(method="ffill", inplace   =False) #substitui os dados nan da coluna salario pelo valor anterior (method="ffill") e altera o dataframe original (inplace=True)
print(df) # retorna o dataframe com os dados nan da coluna salario substituidos pelo valor anterior
#pega o valor de cima e substitui a ausencia
df["Idade"] = df["Idade"].ffill()

#debaixo para cima
df["Idade"] = df["Idade"].bfill()
```

```python
#criar uma nova coluna: faixa etaria
df["Faixa Etária"] = pd.cut(df["Idade"], bins=[0, 18, 30, 60, 100], labels=["Criança", "Jovem", "Adulto", "Idoso"]) #cria uma nova coluna "Faixa Etária" com base na coluna "Idade" utilizando a função cut() do pandas para categorizar as idades em faixas etárias específicas
print(df) # retorna o dataframe com a nova coluna "Faixa Etária" criada com base na coluna "Idade"
```

```python
#remover duplicatas
df = df.drop_duplicates() # remove as linhas duplicadas do dataframe 
print(df) # retorna o dataframe sem as linhas duplicadas   
```

```python
# montar categoria de salario
#df["faixa salarial"] = df["Sálario"].map(lambda x: "Baixo" if x < 2000 else ("media" if x < 3000 else "alto")) # cria uma nova coluna "faixa salarial" com base na coluna "Salario" utilizando a função map() do pandas para categorizar os salários em faixas salariais específicas    
df["faixa salarial"] = df["Salário"].apply(lambda x: "Baixo" if x < 2000 else ("media" if x < 3000 else "alto")) # cria uma nova coluna "faixa salarial" com base na coluna "Salário" utilizando a função apply() do pandas para categorizar os salários em faixas salariais específicas    
print(df)
```

```python
df.to_excel("dados_tratados.xlsx", index=False) # salva o dataframe tratado em um novo arquivo excel chamado "dados_tratados.xlsx" sem incluir o índice (index=False)
```
