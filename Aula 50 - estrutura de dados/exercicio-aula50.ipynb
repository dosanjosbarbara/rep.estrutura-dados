```python
import pandas as pd
import openpyxl as op 

df = pd.read_excel("dados_a_tratar.xlsx")
print(df)
```

```python
#1 - Verificar se há valores faltantes
print(df.isnull().sum())    
```

```python
# 2 - Preencher valores nulos com diferentes estratégias
# Colunas categóricas: substitua por "Desconhecido"
df["Nome"] = df["Nome"].fillna("Desconhecido")
df["Departamento"] = df["Departamento"].fillna("Desconhecido")

# Colunas numéricas: converter para numérico antes de calcular média
df["Idade"] = pd.to_numeric(df["Idade"], errors='coerce')
df["Salário"] = pd.to_numeric(df["Salário"], errors='coerce')

# Calcular a média GERAL de cada coluna (com skipna=True, ignora NaN)
media_idade = df["Idade"].mean()
media_salario = df["Salário"].mean()

# Preencher APENAS os valores nulos com a média geral
df["Idade"] = df["Idade"].fillna(media_idade)
df["Salário"] = df["Salário"].fillna(media_salario)

# Data_Admissao: substitua por "2000-01-01"
df["Data_Admissao"] = df["Data_Admissao"].fillna("2000-01-01")

print("DataFrame após preenchimento de valores nulos com médias:")
print(f"Média de Idade usada: {media_idade}")
print(f"Média de Salário usada: {media_salario}")
print(df)

```

```python
#3.0 - Verificar se há valores duplicados
#df.duplicated().sum()
#print(df.duplicated().sum())

# 3.0 - Visualizar as linhas duplicadas
print("Linhas duplicadas:")
print(df[df.duplicated(keep=False)].sort_values(by=df.columns.tolist()))
```

```python
# 3.1 - Remover linha duplicada (índice 10 = linha 11)
df = df.drop(10, axis=0)
print("DataFrame após remover duplicata:")
print(df)
print(f"\nTotal de linhas: {len(df)}")
```

```python
# 4 - Corrigir e-mail específico do henrique
# Localizar a linha onde Nome é "henrique" e corrigir o Email
df.loc[df['Nome'] == 'Henrique', 'Email'] = 'henrique@email.com'
df.loc[df['Nome'] == 'Fernando', 'Email'] = 'fernando@email.com'
print(df)
```

```python
# 5 - Converter coluna "Data_Admissao" para formato datetime
df["Data_Admissao"] = pd.to_datetime(df["Data_Admissao"], errors='coerce')
print(df)

```

```python
# 6 - Garantir tipos de dados corretos
df["Idade"] = df["Idade"].astype(int)
df["Salário"] = df["Salário"].astype(float)
df["Data_Admissao"] = pd.to_datetime(df["Data_Admissao"])

print(df.dtypes)
print(df)
```

```python
# Formatar exibição dos salários com 4 casas decimais 
df["Salário"] = df["Salário"].round(4)
print(df)
```

```python
# Data Frame 
# 1 quantidade de funcionario por departamento
df_funcionario = df["Departamento"].value_counts()
print(df_funcionario)
```
