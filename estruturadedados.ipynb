```python
import pandas as pd
dados = {
    "nome": ["João", "Carlos", "Maria", "Ana", "Pedro"],
    "idade": [25, 30, 22, 28, 35],
    "Naturalidade": ["SP", "RJ", "MG", "RS", "BA"]
}

df = pd.DataFrame(dados)
print(df)
```

```python
#verificar coluna especifica
print (df["idade"])
```

```python
#puxar valor condicional

print(df["idade"]>30)
```

```python
#puxar valor conditional com informacoes completas
# 1 filtrar idade - df["idade"]>30
# 2 puxar todas as infos - df[]
print(df[df["idade"]<30])
```

```python
# puxar registros de um unico nome

#traz informacoes vazias, indicando apenas atrves do indice quem é verdadeiro para a declaracao =='Ana'
#print(df['nome']=='Ana')
print(df[df['nome']=='Ana')]

```

```python
#estatistica descritiva
df.describe()
```

```python
#filtro de primeiras linhas
df.head(2)
```

```python
#aplicar uma funcao para calcular idade + 2
df['idade2'] = df['idade'].apply(lambda x: x+2) ## ao usar lambda, cada valor (x) da coluna 'idade' recebe a soma de + 2

print(df)
```

```python
# contagem de valor
df['Naturalidade'].value_counts() #retorna a quantidade em que a naturalidade aparece na variavel df
```

```python
#ordenar dados
#asscending = true (ordem crescente)
#ascending = false (ordem descrescente)

df.sort_values("Naturalidade", ascending=False) #ordenar por naturalidade
```

```python
#agrupar por (groupby) / calculo de media de idade por naturalidade
media_nat = df.groupby("Naturalidade")["idade"].mean()
print(media_nat)
```
